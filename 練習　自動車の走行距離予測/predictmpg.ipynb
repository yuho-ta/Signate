```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression as LR
from sklearn.metrics import mean_squared_error as MSE

# データの読み込み
train = pd.read_csv('/content/drive/MyDrive/signate/練習　自動車の走行距離予測/train.tsv', sep='\t')
test = pd.read_csv('/content/drive/MyDrive/signate/練習　自動車の走行距離予測/test.tsv', sep='\t')
samplesubmission = pd.read_csv('/content/drive/MyDrive/signate/練習　自動車の走行距離予測/sample_submit.csv',sep='\t')


test.head()
```

```python
# 欠損値を含む行の削除
train = train.dropna()
test = test.dropna()

#目的変数及び説明変数を表す変数の準備
y_train = train['mpg'] # 目的変数
X_train = train[['cylinders', 'displacement', 'weight','acceleration','model year', 'origin']] # 説明変数


X_test = test[['cylinders', 'displacement', 'weight','acceleration','model year', 'origin']] # 説明変数


X_train.dtypes
```

```python
# モデルの箱の準備および学習
lr = LR()
lr.fit(X_train, y_train)
# 予測値の算出
y_test = lr.predict(X_test)
```

```python
test['mpg']=y_test
```

```python
test[['id', 'mpg']].to_csv('/content/drive/MyDrive/signate/練習　自動車の走行距離予測/submit.csv', header=False, index=False)
```
